```python
df
```

```python
_sex=pd.get_dummies(df['Sex'])
_pclass=pd.get_dummies(df['Pclass'])
_embarked=pd.get_dummies(df['Embarked'])
df_cat=pd.concat((_pclass, _sex, _embarked), axis=1)
df_y=df['Survived']
```

```python
df_cat
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(df.loc[:, ['Age']].values)
_age=scaler.transform(df.loc[:, ['Age']].values)
```

```python
scaler=MinMaxScaler()
scaler.fit(df.loc[:, ['SibSp']].values)
_sibsp=scaler.transform(df.loc[:, ['SibSp']].values)
```

```python
scaler=MinMaxScaler()
scaler.fit(df.loc[:, ['Parch']].values)
_parch=scaler.transform(df.loc[:, ['Parch']].values)
```

```python
scaler=MinMaxScaler()
scaler.fit(df.loc[:, ['Fare']].values)
_fare=scaler.transform(df.loc[:, ['Fare']].values)
```

```python
_num=np.concatenate((_age, _sibsp, _parch, _fare), axis=1)
df_num=pd.DataFrame(_num)
df_num.columns=['Age', 'SibSp', 'Parch', 'Fare']
df_num
```

```python
df_x=pd.concat((df_cat, df_num), axis=1)
df_x
```

```python
df_y
```

```python
from sklearn.svm import SVC
X_np=df_x.values
Y_np=df_y.values.reshape(-1, 1)

model=SVC(random_state=0)
model.fit(X_np, Y_np)

pred=model.predict(X_np)
# "Training data vs Validation data vs Testing data"
# 0~99 => 0~79(Train) / 80~89 (Valid) / 90~99 (Test)
# 0~99 => 0~79(Train) / 0~79 (Valid) / 90~99 (Test)

pred[:10]
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_np, pred)
```
